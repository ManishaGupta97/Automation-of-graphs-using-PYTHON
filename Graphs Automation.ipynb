## Creating a function that will automatically save the graphs of all the columns.
If column is numeric then the graphs will be saved as histogram and boxplots , if columns are categorical then bar plots.

```python
def graphs(data,cols = list(data.columns), directory = (os.getcwd())):
    data = data[cols] # Changing dataframe to the given columns only
    
    for i in range(len(data.columns)):
        if data.iloc[:,i].dtypes == 'object': # Checking if the column is categorical or not
            
            # Plotting barplot for categorical columns
            plt.subplot(1,1,1)
            data[cols[i]].value_counts().plot(kind = 'bar', figsize=(4,4),color="coral",fontsize=13,edgecolor='black')
            plt.xlabel(cols[i],fontsize=16,color='green')
            plt.ylabel("Frequency",fontsize=16,color='green')
            plt.title(f"Barplot of {cols[i]}",fontsize=20,color='green')
                       
        else:
            
            # Plotting histogram for numerical values in subplot 1
            sns.set(font_scale=2)
            plt.figure(figsize=(20,5))
            plt.subplot(1,2,1)
            sns.histplot(data ,x = cols[i],color = 'coral').set_title(f"Histogram of {cols[i]}",fontsize=20,color='Green')
            
            # Plotting boxplot for numerical column in subplot 2
            plt.subplot(1,2,2)
            sns.boxplot(x = data[cols[i]],color='pink').set_title(f"Boxplot of {cols[i]}",fontsize=20,color='green') # Specifying column, colour and title for the boxplot
            
        direct = directory+"/"+cols[i]+'_'+"plot.png" # Adding name of the figure to the directory
        plt.savefig(direct,bbox_inches='tight') # Saving the plot in the specified path
```

## Calling the function

If the directory is not mentioned then the graphs will be automatically saved in the current working directory.
Similarly if we do not specify the columns, graphs of all columns will be plotted and saved.

```python
graphs(data,cols=['Acceleration', 'Model', 'Origin', 'Horsepower', 'Cylinders'])
```
